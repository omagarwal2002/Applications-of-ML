EM algorithm and K-means for clustering and comparing them

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import accuracy_score, adjusted_rand_score
```

```python
# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
kmeans_labels = kmeans.labels_
```

```python
# Compute the accuracy and adjusted rand score for K-means
kmeans_acc = accuracy_score(y, kmeans_labels)
kmeans_ars = adjusted_rand_score(y, kmeans_labels)
```

```python
# Apply EM clustering
em = GaussianMixture(n_components=3, random_state=42)
em.fit(X)
em_labels = em.predict(X)
```

```python
# Compute the accuracy and adjusted rand score for EM
em_acc = accuracy_score(y, em_labels)
em_ars = adjusted_rand_score(y, em_labels)
```

```python
# Plot the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

ax1.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
ax1.set_title(f'K-means\nAccuracy: {kmeans_acc:.2f}, Adjusted Rand Score: {kmeans_ars:.2f}')

ax2.scatter(X[:, 0], X[:, 1], c=em_labels, cmap='viridis')
ax2.set_title(f'EM\nAccuracy: {em_acc:.2f}, Adjusted Rand Score: {em_ars:.2f}')

plt.show()
```
